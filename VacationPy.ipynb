### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
#Configure GMAPS
gmaps.configure(API_KEY)
```

```python
humidity=vacationdf['humidity'].astype(float)
max_humidity=humidity.max()
city=vacationdf[['lattitude','longitude']]
fig=gmaps.figure(center = [31.9686, -99.9018], zoom_level = 3)
heat_layer=gmaps.heatmap_layer(city,weights=humidity,max_intensity=max_humidity,dissipating=False,point_radius=2)
fig.add_layer(heat_layer)
fig
```

```python
hlist=len(set(best1['Unnamed: 0']))
hlist
```

```python
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
params = {
    'type':"lodging",
    "keyword":"hotel",
    "radius": 20000,
    "key": API_KEY,
}
  
```

```python
for  index,row in best1.iterrows():
    lat=row['lattitude']
    lng=row['longitude']
    city=row['city']
    params['location']=f'{lat},{lng}'
    print (f'retriving {city}')
    
    response=requests.get(base_url,params=params).json()
    results=response['results']
    
    try:
        best1.loc[index,'hotel']=results[0]['name']
    except (KeyError,IndexError):
        print("missing")
```
